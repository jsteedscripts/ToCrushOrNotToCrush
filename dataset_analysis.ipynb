```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
climbers = pd.read_csv("archive(4)/climber_df.csv")
conversions = pd.read_csv("archive(4)/grades_conversion_table.csv")
routes = pd.read_csv("archive(4)/routes_rated.csv")
```

```python
def cat(a):
    if a <= 44:
        return "Intermediate"
    elif a > 44 and a <= 57:
        return "Advanced"
    else:
        return "Expert"
```

```python
features = ['grades_first', 'grades_max', 'grades_count', 'grades_mean']

sub = climbers[features]

sns.pairplot(sub)
plt.savefig('linear_features_pairplot.png')
plt.show()
```

```python
climbers['BMI'] = (climbers['weight'] / climbers['height'] / climbers['height']) * 10000
print(climbers.describe())
```

```python
plt.figure(figsize=(8, 6))

climbers.boxplot(column='BMI')
# plt.savefig('grades_count_box_plot.png')
plt.title(f'BMI Box Plot')
plt.show()
```

```python
std_devs = climbers[['height', 'weight', 'age', 'years_cl', 'grades_count', 'grades_max', 'grades_mean', 'grades_first', 'BMI']].std()

plt.figure(figsize=(12, 6))
sns.barplot(x=std_devs.index, y=std_devs.values, color='lightcoral')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Features')
plt.ylabel('Standard Deviation')
plt.title('Standard Deviations of Features')
plt.savefig('stddev_features.png')
plt.show()
```

```python
features = ['height', 'weight', 'age', 'years_cl', 'grades_count', 'grades_max', 'grades_mean', 'grades_first'] 

sns.set(style="whitegrid")

fig, ax = plt.subplots()

for feature in features:
    sns.kdeplot(climbers[feature], fill=True, label=feature)

plt.xlabel("Feature Values")
plt.ylabel("Density")
plt.title("Density Plot of Climber Features")

ax.legend()
plt.xlim(xmin=0, xmax=300)
plt.savefig('density_plot.png')
plt.show()
```

```python
correlation_matrix = climbers[features].corr()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.savefig('correlation_matrix_heatmap.png')
plt.show()
```
